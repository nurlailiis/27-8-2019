### Boxplot

```python
join_array = np.array(join[['city-mpg','highway-mpg']])
```

```python
colors = ['blue', 'grey', 'lawngreen']

bp = plt.boxplot(join_array, 
                 patch_artist=True,
                 notch=True)

for i in range(len(bp['boxes'])):
    
    bp['boxes'][i].set(facecolor=colors[i])
    
    bp['caps'][2*i + 1].set(color=colors[i])
    
plt.xticks([1, 2], ['city-mpg','highway-mpg'])

plt.show()
```

### Bar Plot

```python
#filter by full-type=gas
data_gas = join[join['fuel-type']=='gas']
top_10_g=data_gas.sort_values('curb-weight',ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))

x=range(10)
plt.bar(x,top_10_g['curb-weight']/10**9)
plt.xticks(x,top_10_g['make'])
plt.xlabel('Type of Cars')
plt.ylabel('Curb Weight')
plt.title('10 Most Fuel Type In Cars')
plt.show()
```

### Scatter Plot

```python
plt.scatter(data_gas['city-mpg'],data_gas['highway-mpg'])
plt.title('Combined Driving and Highway Driving Miles per Gallon in Vehicles')
plt.xlabel('Combined Driving')
plt.ylabel('Highway Driving')
plt.show()
```

### Pie Chart

```python
joins = join[['fuel-system','engine-size']]
joins.head()
```

```python
join_group = joins.groupby('fuel-system').mean()
join_group.head()
```

```python
plt.figure(figsize=(8,8))
plt.pie(join_group, 
        labels=join_group.index, autopct='%.2f')

plt.axis('equal')
plt.show()
```

### Violin

```python
vp = plt.violinplot(join_array,
                    showmedians=True)

plt.xticks([1, 2], ['city-mpg','highway-mpg'])

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor='darkblue')

plt.show()    
```

### Doughnut Chart

```python
# Create a circle for the center of the plot
plt.figure(figsize=(8,8))
plt.axis('equal')
my_circle=plt.Circle( (0,0), 0.7, color='white')
my_circle
# Give color names
plt.pie(join_group, labels=join_group.index, colors=['red','green','blue','skyblue'], autopct='%.2f')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

## PART 2

#### 1. Car brand with highest price 

```python
carbrand = join.groupby('make').price.max().reset_index()
data_sort = carbrand.sort_values(by='price', ascending=False).reset_index()
data_sort.drop(columns='index').head(1)
```

#### 2. Maximum horsepower for car with 6 cylinder engine 

```python
max_horse = join[join['cylinder']=='six']
data_max = max_horse['horsepower'].max()
data_max
```

#### 3. Average peak-rpm for â€˜turboâ€™ style aspiration car

```python
avg1 = join[join['aspiration']=='turbo']
data_avg = avg1['peak-rpm'].mean()
data_avg
```

#### 4. Average price based on body-style

```python
caravg1 = join.groupby('body-style').price.mean().sort_values().reset_index()
caravg1
```
